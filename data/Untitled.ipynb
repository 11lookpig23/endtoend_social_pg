```python
import numpy as np
import matplotlib.pyplot as plt

social = np.load("pg_social.npy")
plt.plot([i*2e4 for i in range(50)],[sum(social[i*40:(i+1)*40])/40 for i in range(50)])#(range(len(social)),social)#(range(50),[sum(acdata[i*40:(i+1)*40])/40 for i in range(50)])
plt.savefig("social_law.png")
plt.xlabel("steps")
plt.ylabel("reward")
plt.title('social-law')
plt.show()
```
